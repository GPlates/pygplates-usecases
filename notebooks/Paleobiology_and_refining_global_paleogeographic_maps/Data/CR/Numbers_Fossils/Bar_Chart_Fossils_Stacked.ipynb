```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#read fossil data
Fossil_data=pd.read_csv('/Users/wenchaocao/Research/5_Consistency_Ratios_Fossils_Paleogeography/Numbers_Fossils_Python/Consistency_Ratio_RotationCorrections_latest_Simplified.csv', delimiter=',')  
Fossil_data
```

```python
bar_width = Fossil_data['bar_width']

# bar diagram
plt.figure(figsize=(12,8),tight_layout=True)
p2 = plt.bar(Fossil_data['Ave_age'], Fossil_data['Consistent Marine Fossils']+Fossil_data['Inconsistent Marine Fossils']+Fossil_data['Consistent Terrestrial Fossils']+Fossil_data['Inconsistent Terrestrial Fossils'], bar_width,align='center', color='b') 
p1 = plt.bar(Fossil_data['Ave_age'], Fossil_data['Consistent Terrestrial Fossils']+Fossil_data['Inconsistent Terrestrial Fossils'], bar_width, align='center',color='r')

# x, y labels, ticks
plt.xlabel('Time(Ma)',fontsize=20)
plt.ylabel('Numbers of Fossil Collections',fontsize=20)
plt.legend((p1[0], p2[0]), ('Terrestrial Fossils', 'Marine Fossils'),fontsize=12)

# Label the x axis at the boundaries between the different stages
plt.xticks(np.hstack((Fossil_data['From_age'],Fossil_data['To_age'])), size=12, rotation=50)
plt.xlim(402,2)
plt.yticks(np.arange(0, 5001, 500),size=12)
plt.grid(axis='y') #color='r',linestyle=':', linewidth=2
plt.savefig('FossilsNumbers_Stacked_BarChart.ps',dpi=600)
plt.savefig('FossilsNumbers_Stacked_BarChart.pdf',dpi=600)

plt.show()

print np.hstack((Fossil_data['From_age'],Fossil_data['To_age']))
```
